```python
def auto_number():
    global df2
    wbs_ = input("WBS: ").split(';')
    number_ = input("Number Value: ")
    code_ = input("Code Name: ")    
    pre_result = []
    for i in df2.index:
        try:
            item = df2.iloc[i]
            _level = item[header_wbs_level]
            for sw in wbs_:
                try:
                    if str(item[header_wbs]).startswith(sw):
                        pre_result.append(f"{_level}\t{number_}\t{df2.iloc[i][header_wbs]}\t\t{df2.iloc[i][header_task_name]}")
                        df2[code_][i] = number_
                except:
                    pass
        except:
            pass
    [print(s) for s in pre_result[:10]]
```
