```python
header_key = 'Work Item Number'
data = df[header_key]# print(data)
data
```

```python
data_list = [str(d).split("-") for d in data]
[d for d in data_list]
```

```python
data_level = [len(d) for d in data_list] #[d for d in data_level]
```

```python
dic_level = {} # tự điển các Level 
for d in data_level:
    dic_level[d] = []
for d in data_list:
    dic_level[len(d)].append(d)

[f"{d} : {dic_level[d]}" for d in dic_level]
```

```python
dic_code = abserviation() # Tự điển tra cứu mã số -> Tên
[f"{d} : {dic_code[d]}" for d in dic_code]
```

```python
parent_dic_list = []
for d in data_list:
    if len(d) > 1:
        parent = d[:-1]
        if not parent in dic_level[len(parent)]:
            dic = {}
            dic["Work Item Number"]  = "-".join(d[:-1])
            dic["Work Item Name"]= " ".join([dic_code[c] for c in parent if c in dic_code])
            parent_dic_list.append(dic)
parent_dic_list
```

```python
dic_add = {} # tự điển các Mã CT Parent cần add thêm
for d in data_list:
    if len(d) > 1:
        parent = d[:-1]
        if not parent in dic_level[len(parent)]:
            dic_add["-".join(d[:-1])] = ""
# [f"{d}" for d in dic_add]
for d in dic_add:
    codes = d.split("-")
    dic_add[d] = " ".join([dic_code[c] for c in codes if c in dic_code])
[f"{d}:{dic_add[d]}" for d in dic_add]
```

```python
# combine Start end for Additional Parent Items
"""'Work Item Number',
 'Work Item Name',
 'Plan Start Date',
 'Plan End Date',
 'Actual Start Date',
 'Acutal End Date / Actual Tracking date',
 'PIC',
 '#NCR',
 '% Complete',
 'Report Date',
 'Quantity Unit',
 'BIM Quantity',
 'QS Quantity',
 'Cost'"""
dic_start = {}
dic_end = {}
dic_duration = {}
for i in index:
    try:
        wi = df.iloc[i]
        parent = "-".join(wi["Work Item Number"].split("-")[:-1])
        start = df.iloc[i]["Actual Start Date"]
        end = df.iloc[i]["Acutal End Date / Actual Tracking date"]
        if parent in dic_add:
            if not parent in dic_start:            
                dic_start[parent] = start            
            if not parent in dic_end:            
                dic_end[parent] = end
            if start <  dic_start[parent]:
                dic_start[parent] = start 
            if end > dic_end[parent]:            
                dic_end[parent] = end

            dic_duration[parent] = (dic_end[parent] - dic_start[parent]).days+1
        
    except Exception as ex:
        pass  
```
