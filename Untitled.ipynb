```python
data_level = [len(d) for d in data_list] #[d for d in data_level]
```

```python
dic_level = {} # tự điển các Level 
for d in data_level:
    dic_level[d] = []
for d in data_list:
    dic_level[len(d)].append(d)

[f"{d} : {dic_level[d]}" for d in dic_level]
```

```python
dic_code = abserviation() # Tự điển tra cứu mã số -> Tên
[f"{d} : {dic_code[d]}" for d in dic_code]
```

```python
dic_add = {} # tự điển các Mã CT Parent cần add thêm
for d in data_list:
    if len(d) > 1:
        parent = d[:-1]
        if not parent in dic_level[len(parent)]:
            dic_add["-".join(d[:-1])] = ""
# [f"{d}" for d in dic_add]
for d in dic_add:
    codes = d.split("-")
    dic_add[d] = " ".join([dic_code[c] for c in codes if c in dic_code])
[f"{d}:{dic_add[d]}" for d in dic_add]
```

```python
# combine Start end for Additional Parent Items
"""'Work Item Number',
 'Work Item Name',
 'Plan Start Date',
 'Plan End Date',
 'Actual Start Date',
 'Actual End Date',
 'PIC',
 '#NCR',
 '% Complete',
 'Report Date',
 'Quantity Unit',
 'BIM Quantity',
 'QS Quantity',
 'Cost'"""

dic_comb_list = []
dic_time = {}
for i in index:    
    try:
        wi = df.iloc[i]
        parent_numbers = wi["Work Item Number"].split("-")[:-1]
        parent = "-".join(parent_numbers)
        start = df.iloc[i]["Plan Start Date"]
        end = df.iloc[i]["Plan End Date"]
        start_actual = df.iloc[i]["Actual Start Date"]
        end_actual = df.iloc[i]["Actual End Date"]              
        if parent in dic_add:
            if not parent in dic_time:
                dic_time[parent] = [[],[],[],[]]
            else:
                dic_time[parent][0].append(start)
                dic_time[parent][1].append(end) 
                dic_time[parent][2].append(start_actual) 
                dic_time[parent][3].append(end_actual)   
    except Exception as ex:
        pass  
debugger = []
for d in dic_time:
    try:
        dic = {}
        dic["Work Item Number"] = d
        dic["Work Item Name"] = " ".join([dic_code[c] for c in d.split("-") if c in dic_code])
        try:
            dic["Plan Start Date"] = min(dic_time[d][0])
        except:
            dic["Plan Start Date"] = None
        try:
            dic["Plan End Date"] = max(dic_time[d][1])
        except:
            dic["Plan End Date"] = None
        try:
            dic["Actual Start Date"] = min(dic_time[d][2])
        except:
            dic["Actual Start Date"] = None
        try:
            dic["Actual End Date"] = max(dic_time[d][3])  
        except:
            dic["Actual End Date"] = None       
        
        
        dic_comb_list.append(dic)
    except Exception as ex:
        debugger.append(ex.args[0])
        pass
dic_comb_list
```
