#### EXPLORE DATA

```python
def level(str_):
    level = len(str_.split("."))-1
    return level
```

```python
index = df.index # range
columns = df.columns
col_count = len(columns)
columns
```

```python
target_level = 4
for i in index:
    try:
        lev = level(str(df.iloc[i]["WBS"]))
        if lev == target_level:
            print(f"{target_level}\t{df.iloc[i]['WBS']}\t\t{df.iloc[i]['Task Name']}")
    except Exception as ex:
        print(ex)
        pass    
```

```python
_level = 3
for i in index:
    try:
        lev = level(str(df.iloc[i]["WBS"]))
        if lev <= _level:
            print(f"{lev}\t{df.iloc[i]['WBS']}\t\t{df.iloc[i]['Task Name']}")
    except Exception as ex:
        print(ex)
        pass   
```

```python
targer_wbs = input("WBS number: ")
for i in index:
    try:
        lev = level(str(df.iloc[i]["WBS"]))
        wbs = str(df.iloc[i]["WBS"])
        if wbs.startswith(targer_wbs):
            print(f"{lev}\t{df.iloc[i]['WBS']}\t\t{df.iloc[i]['Task Name']}")
    except Exception as ex:
        print(ex)
        pass   
```

```python
memory = {"FD":[],
         "CR":[],
         "CL":[]}
```

```python
# EXPLORE KEYWORD
target_keyword = unidecode.unidecode(input("Keyword: ").lower())
for i in index:
    try:
        lev = level(str(df.iloc[i]["WBS"]))
        taskname = unidecode.unidecode(str(df.iloc[i]["Task Name"]).lower())
        if target_keyword in taskname:
            print(f"{lev}\t{df.iloc[i]['WBS']}\t\t{df.iloc[i]['Task Name']}")
    except Exception as ex:
        print(ex)
        pass  
```

```python
# TRAIN MEMORY
while True:
    memo_key = input("KEY? : ").upper()
    if memo_key.lower() == "qqq":
        break
    value_keyword = unidecode.unidecode(input("Keyword: ").lower())
    if not memo_key in memory:   
        memory[memo_key] = []
    memory[memo_key].append(value_keyword)    
    
```

```python
memory
```

```python
memo_df = pd.DataFrame(Series(memory),columns = [["Memory Value"]])
path_excel_memo = r"K:\_WFH VIETMY\XLSX\MEMORY.xlsx"
memo_df.to_excel(path_excel_memo)
```
